## Stacked Area Chart

Create a stacked area chart of a printer company and its revenue for the last 12 months. The data set contains monthly values in the span of a year and features three revenue sources:
-	International Printer Sales
-	Domestic Printer Sales
-	Services (fixed fee)



```python
revenue2020 = pd.read_csv('area_chart_homework_data.csv')
revenue2020.head()
```

```python
revenue_color = ['#0C7489', '#119DA4', '#FF4000']
revenue_labels = ['International', 'Domestic', 'Services']
sns.set_style('white')

plt.figure(figsize = (15,8))
plt.stackplot(revenue2020['Date'],
              revenue2020['International'],
              revenue2020['Domestic'],
              revenue2020['Services'],
              colors = revenue_color,
              edgecolor = 'none')

sns.despine()
plt.xticks(revenue2020['Date'], rotation = 45)
plt.ylabel('Revenue')
plt.legend(labels = revenue_labels, loc = 'upper left')
plt.title('Revenue Report 2020', fontsize = 14, weight = 'bold')
plt.show()
```

## Line Chart

Create a line chart using the S&P 500 and FTSE 100 index data from 2010 until 2020.
After created the line chart:
- Narrowing down the time frame to a more specific period.
- Create another line chart based on the specified period.

```python
sp500_ftse1020 = pd.read_csv('returns1020.csv')
sp500_ftse1020
```

```python
#change value into date format
sp500_ftse1020['new_date'] = pd.to_datetime(sp500_ftse1020['Date'])
sp500_ftse1020['new_date']
```

```python
index_labels = ['S&P 500', 'FTSE 100']

plt.figure(figsize = (20,8))
plt.plot(sp500_ftse1020['new_date'], sp500_ftse1020['GSPCRet'], color = '#BB0A21')
plt.plot(sp500_ftse1020['new_date'], sp500_ftse1020['FTSERet'], color = '#4B88A2')

plt.legend(labels = index_labels, loc = 'upper right')
plt.show()
```

```python
#specify to 2011 timeframe
sp500_ftse_2011 = sp500_ftse1020[(sp500_ftse1020['new_date'] >= '2011-01-01') &
                                 (sp500_ftse1020['new_date'] <= '2011-12-31')]
sp500_ftse_2011
```

```python
plt.figure(figsize = (20,8))
plt.plot(sp500_ftse_2011['new_date'], sp500_ftse_2011['GSPCRet'], color = '#EE4266')
plt.plot(sp500_ftse_2011['new_date'], sp500_ftse_2011['FTSERet'], color = '#540D6E')

plt.legend(labels = index_labels, loc = 'upper right')
plt.title('S&P 500 vs FTSE 100 Returns 2011', fontsize = 16, weight = 'bold')
plt.show()
```

## Histogram

The dataset contains a survey of geodemographic features among 2000 participants. Create a histogram based on the Age variable in the data.

```python
cust_survey = pd.read_csv('histogram_survey_data.csv')
cust_survey
```

```python
sns.set_style

plt.figure(figsize = (8,6))
plt.hist(cust_survey['Age'],
         bins = 12,
         color = '#61E786')

plt.xlabel('Age')
plt.xticks(np.arange(min(cust_survey['Age']), max(cust_survey['Age']), 5))
plt.ylabel('Number of Participants')
sns.despine()
plt.title('Customer Survey Age Distribution', fontsize = 14, weight = 'bold')
plt.show()
```

## Regression Plot

Create a regression plot on the dataset of car insurance in Sweden.

```python
car_insurance = pd.read_csv('regression_auto_insurance_sweden.csv')
car_insurance
```

```python
sns.set_style('darkgrid')

plt.figure(figsize = (10,7))
sns.regplot(x = car_insurance['Claims'],
            y = car_insurance['Payment'],
            data = car_insurance,
            scatter_kws = {'color':'#1E555C'},
            line_kws = {'color':'#C1292E'})

plt.title('Car Insurance Claims and Payment in Sweden', fontsize = 14, weight = 'bold')
plt.show()
```

## Bar and Line Chart

Create a Pareto chart. A Pareto is a specific case of a bar and line chart. The dataset contains the top 5 Movie Goers complaints. Your task is to create a Pareto chart based on the data.

```python
movie_complaints = pd.read_csv('bar_line_chart_homework.csv')
movie_complaints
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.bar(movie_complaints['Movie Theater Goers Complaints'],
       movie_complaints['Number of complaints'],
       color = '#1E555C')
ax.set_ylabel('Number of Complaints', weight = 'bold')
ax.tick_params(axis = 'y', width = 2, labelsize = 'large')

ax1 = ax.twinx()
ax1.set_ylim(0, 1.1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(movie_complaints['Movie Theater Goers Complaints'],
         movie_complaints['frequency'],
         color = 'orange',
         marker = 'D')
ax1.set_ylabel('Frequency', weight = 'bold')
ax1.tick_params(axis = 'y', width = 2, labelsize = 'large', color = 'orange')

ax.set_title('Top 5 Movie Goers Complaints Frequency Distribution', fontsize = 14, weight = 'bold')
```
